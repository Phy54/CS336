```python
import torch
import torch.nn.functional as F
def mean_pooling(vector_embedding, attention_mask):
  token_embeddings = vector_embedding[0]
  input_mask_expanded = attention_mask.unsqueeze(-1).expand(token_embeddings.size()).float()
  return torch.sum(token_embeddings* input_mask_expanded,1)/ torch.clamp(input_mask_expanded.sum(1), min=1e-9)
def embedding_function(batch,tokenizer, model_embedding):
  tokenization = tokenizer(batch,padding = True,truncation =True, return_tensors = 'pt')
  with torch.no_grad():
    embedding_vector = model_embedding(**tokenization)
  sentence_embeddings = mean_pooling(embedding_vector,tokenization['attention_mask'])
  sentence_embeddings = F.normalize(sentence_embeddings,p=2,dim=1)
  return [list(sentence_embeddings[0].numpy())]
```

```python
from transformers import AutoModel, AutoTokenizer
berttokenizer = AutoTokenizer.from_pretrained('sentence-transformers/all-MiniLM-L6-v2', cache_dir="/content/drive/MyDrive/KaggleDatasets/embedcache/")
bertmodel = AutoModel.from_pretrained("sentence-transformers/all-MiniLM-L6-v2", cache_dir="/content/drive/MyDrive/KaggleDatasets/embedcache/")
```

```python
unique_books = unique_books.dropna()
```

```python
unique_books.isna().sum()
```

```python
from tqdm import tqdm

embeddings = []
progress_bar = tqdm(total = len(unique_books), desc="Processing")
for idx,row in unique_books.iterrows():
  string = row[1]
  embedding = embedding_function(string,berttokenizer,bertmodel)[0]
  embeddings.append(embedding)
  progress_bar.update(1)
unique_books['vector'] = [row for row in embeddings]
```

```python
from transformers import GPT2LMHeadModel, GPT2Tokenizer

# Tải mô hình GPT-2 và tokenizer
model_name = "gpt2"  # Có thể dùng mô hình nhỏ hơn như gpt2-small
tokenizer = GPT2Tokenizer.from_pretrained(model_name)
model = GPT2LMHeadModel.from_pretrained(model_name)
```

```python
from transformers import T5Tokenizer, T5ForConditionalGeneration

# 1. Load mô hình và tokenizer
tokenizer = T5Tokenizer.from_pretrained("t5-base")
model = T5ForConditionalGeneration.from_pretrained("t5-base")



```

```python
prompt = f"""recommend: Based on the description, suggest the best book.
context: 1. {document_for_LLM[0]} 2. {document_for_LLM[1]} 3. {document_for_LLM[2]} 4. {document_for_LLM[3]} 5. {document_for_LLM[4]}
query: {query}"""
```

```python
inputs = tokenizer(prompt, return_tensors="pt", max_length=512, truncation=True)

# 4. Generate câu trả lời
outputs = model.generate(
    inputs.input_ids,
    max_length=100,
    num_beams=5,
    early_stopping=True
)

# 5. Decode câu trả lời
answer = tokenizer.decode(outputs[0], skip_special_tokens=True)

print("Query:", query)
print("Generated Answer:", answer)
```
